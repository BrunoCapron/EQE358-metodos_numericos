Algoritmo ingenuo para resolução de sistemas de equações e calculo de determinante (sem pivotamento)

```python
import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6],
              [1, 3, 2]])

b = np.array([6, 15, 15])

n = b.size

# implementação da eliminação direta
for k in range(n - 1):
    for i in range(k + 1, n):
      m = A[i, k] / A[k, k]
      for j in range(k, n):
        A[i, j] = A[i, j] -  A[k, j] * m
      b[i] = b[i] -  b[k] * m

# cálculo do determinante de A
det = 1
for i in range(n):
  det = det * A[i, i]
print('O determinante de A é:')
print(det)

# implementação da substituição reversa
x = np.zeros(n) # inicialização do vetor de soluções
x[n - 1] = b[n - 1] / A[n - 1, n - 1]
for i in range(n - 2, -1, -1):
    s = b[i]
    for j in range(i + 1, n):
        s = s - A[i, j] * x[j]
    x[i] = s / A[i, i]
print('a solução do sistema é:')
print(x)

# resolução usando o solver do numpy
# A = np.array([[1, 2, 3],
#               [4, 5, 6],
#               [1, 3, 2]])
# b = np.array([6, 15, 15])
# print('Com o solver do numpy:')
# x = np.linalg.solve(A, b)
# print(x)

```
