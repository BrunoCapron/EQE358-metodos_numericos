Algoritmo para resolução de sistemas de equações lineares e calculo de determinante (com pivotamento)

```python
import numpy as np

A = np.array([[1, 1, 1],
              [1, 1, 2],
              [1, 2, 2]])

b = np.array([1, 2, 1])

n = b.size

num_pivots = 0
# implementação da eliminação direta
for k in range(n - 1):
  p=k
  maior = abs(A[k, k])
  for i in range(k + 1, n):
    if abs(A[i, k]) > maior:
      maior = abs(A[i, k])
      p = i
  if p != k:
    num_pivots = num_pivots + 1
    for j in range(k, n):
      aux = A[p, j]
      A[p, j] = A[k, j]
      A[k, j] = aux
    aux = b[p]
    b[p] = b[k]
    b[k] = aux
  for i in range(k + 1, n):
    m = A[i, k] / A[k, k]
    for j in range(k, n):
      A[i, j] = A[i, j] -  A[k, j] * m
    b[i] = b[i] -  b[k] * m

# cálculo do determinante de A
det = (-1)**num_pivots
for i in range(n):
  det = det * A[i, i]
print('O determinante de A é:')
print(det)

# implementação da substituição reversa
x = np.zeros(n) # inicialização do vetor de soluções
x[n - 1] = b[n - 1] / A[n - 1, n - 1]
for i in range(n - 2, -1, -1):
    s = b[i]
    for j in range(i + 1, n):
        s = s - A[i, j] * x[j]
    x[i] = s / A[i, i]
print('a solução do sistema é:')
print(x)

# # resolução usando o solver do numpy
# A = np.array([[1, 1, 1],
#               [1, 1, 2],
#               [1, 2, 2]])
# b = np.array([1, 2, 1])
# print('Com o solver do numpy:')
# x = np.linalg.solve(A, b)
# print(x)
# print('O determinante de A é:')
# print(np.linalg.det(A))

```
