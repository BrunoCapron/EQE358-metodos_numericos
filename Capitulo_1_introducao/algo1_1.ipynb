Transformação de um número inteiro N em base decimal para a base binária




````python
import math

def decToBin(N): # Define uma função para converter decimal para binário.

  # Identificação da maior potência de 2 contida em N
  # Determinar N tal que 2^n <= N < 2^(n+1)
  n=0
  pot=1 #2^0
  while pot <= N:
    n+=1
    pot*=2

  pot=pot/2
  n-=1

  # Calculo dos coeficientes ai em base binária

  Bin = [] #an
  M = N
  for i in range(n+1):
    if M <pot:
      Bin.append(0)
    else:
      Bin.append(1)
      M = M - pot
    pot=pot/2
  return Bin

N=15
Bin = decToBin(N)
print('A representação binária de '+ str(N) + ' é:'+ str(Bin))
````

# 🧮 Convertendo um número decimal para binário com Python

Vamos aprender passo a passo como funciona esse código Python que converte um número decimal (ex: 15) para binário (como 1111).

---

## 🧠 Objetivo

Transformar um número que usamos no dia a dia (decimal) para a forma como os computadores entendem (binário).

---

## 💡 Conceito básico de binário

No sistema binário, usamos **apenas 0 e 1**.
Cada posição representa uma **potência de 2**.
Por exemplo, o número decimal 15 em binário é:

```
2³  2²  2¹  2⁰
 8 + 4 + 2 + 1 = 15
 →   1   1   1   1
```

Portanto, 15 em binário é `1111`.

---

## 🧾 O código Python


```python
import math

def decToBin(N):
    # Descobrir a maior potência de 2 menor ou igual a N
    n = 0
    pot = 1  # Começa com 2^0

    while pot <= N:
        n += 1
        pot *= 2  # Multiplica por 2 em cada passo

    pot = pot / 2
    n -= 1

    # Construir a representação binária
    Bin = []  # Lista que vai guardar os 0s e 1s
    M = N     # Número original, que será diminuído

    for i in range(n + 1):
        if M < pot:
            Bin.append(0)
        else:
            Bin.append(1)
            M = M - pot
        pot = pot / 2  # Vai para a próxima potência menor

    return Bin

# Testando a função
N = 15
Bin = decToBin(N)
print('A representação binária de', N, 'é:', Bin)
```

---

## 🧩 Explicação do código passo a passo

### 1. `import math`

Importa a biblioteca `math`, mas que **não é usada nesse código**.

---

### 2. `def decToBin(N):`

Define uma **função** chamada `decToBin` que recebe um número `N`.

* `def` → palavra-chave para criar funções.
* `N` → o número que queremos converter.

---

### 3. Encontrando a maior potência de 2

```python
n = 0
pot = 1

while pot <= N:
    n += 1
    pot *= 2
```

* Esse loop `while` calcula a **maior potência de 2 que cabe no número N**.
* Exemplo com `N = 15`:

| Repetição | `pot` (2^n)    |
| --------- | -------------- |
| 1         | 1              |
| 2         | 2              |
| 3         | 4              |
| 4         | 8              |
| 5         | 16 (para aqui) |

→ A maior potência de 2 que **não passa** de 15 é `8`, que é `2^3`.

Depois, fazemos:

```python
pot = pot / 2  # volta para 8
n -= 1         # n vira 3
```

---

### 4. Construindo o número binário

```python
Bin = []
M = N
```

* `Bin` é a lista onde vamos guardar os dígitos (0 ou 1).
* `M` é uma cópia de `N`, que vamos diminuir à medida que usamos potências de 2.

---

### 5. Loop para montar os bits

```python
for i in range(n + 1):
    if M < pot:
        Bin.append(0)
    else:
        Bin.append(1)
        M = M - pot
    pot = pot / 2
```

Esse comando significa:

> “Repita os passos abaixo **(n + 1)** vezes”.

* `n + 1` é usado porque estamos contando **inclusive a potência 2⁰**, que também faz parte do número binário.
* Exemplo: Se `n = 3`, o loop será executado com `i = 0, 1, 2, 3` (4 vezes no total).

💡 **`range(n + 1)`** é como criar uma **sequência de números de 0 até n**.

---

### 🔎 Variáveis usadas

Antes do `for`, já temos algumas variáveis importantes:

* `M`: é o número que vai ser representado (vai diminuindo conforme usamos potências de 2).
* `pot`: é o valor da potência de 2 que estamos testando (começa com a maior e vai diminuindo).
* `Bin`: é a lista onde vamos guardar os **0s e 1s** da resposta final.

---

## 👇 Passos dentro do loop

Agora vamos ver o que acontece **em cada repetição** do `for`.

### 🔹 `if M < pot:`

Essa linha faz uma **pergunta**:

> “A potência atual de 2 cabe dentro de `M`?”

Se **não cabe** (ou seja, `M` é menor que `pot`), colocamos `0`:

```python
Bin.append(0)
```

---

### 🔹 `else:` → quando a potência **cabe**

Se a potência cabe no valor que ainda temos (`M`), fazemos 2 coisas:

```python
Bin.append(1)    # Adiciona o número 1 à lista
M = M - pot      # Subtrai essa potência do valor restante
```

Ou seja: usamos essa “potência de 2” e atualizamos o quanto ainda falta.

---

### 🔻 `pot = pot / 2`

No final de cada repetição, dividimos `pot` por 2 para testar a **próxima menor potência de 2**.

Exemplo: se `pot = 8`, depois será `4`, depois `2`, depois `1`, etc.

---

## 🧪 Exemplo (N = 15)

| Passo | Potência | Valor (`pot`) | `M` atual | `M ≥ pot`? | Bit | Cálculo do Novo `M` | Novo `M` |
| ----- | -------- | ------------- | --------- | ---------- | --- | ------------------- | -------- |
| 1     | 2³       | 8             | 15        | Sim        | 1   | `15 - 8`            | 7        |
| 2     | 2²       | 4             | 7         | Sim        | 1   | `7 - 4`             | 3        |
| 3     | 2¹       | 2             | 3         | Sim        | 1   | `3 - 2`             | 1        |
| 4     | 2⁰       | 1             | 1         | Sim        | 1   | `1 - 1`             | 0        |


Resultado final: `Bin = [1, 1, 1, 1]`

---

## 🔚 Resumo final da lógica

1. Começamos com o valor original (ex: 15).
2. Verificamos se a potência de 2 atual **cabe** dentro desse valor.
3. Se sim, colocamos `1` e subtraímos.
4. Se não, colocamos `0`.
5. Repetimos para a próxima potência menor (dividindo por 2).
6. No final, temos uma lista com 0s e 1s representando o número binário.

## 🧪 2º Exemplo: N = 18

Vamos converter o número **18** para binário.

### Passo 1: Encontrar a maior potência de 2 ≤ 18

* `2⁰ = 1`
* `2¹ = 2`
* `2² = 4`
* `2³ = 8`
* `2⁴ = 16`
* `2⁵ = 32` → passou, então usamos até `2⁴ = 16`.

---

### Tabela passo a passo

| Potência | Valor (`pot`) | `M` atual | `M ≥ pot`? | Bit | Cálculo do Novo `M` | Novo `M` |
| -------- | ------------- | --------- | ---------- | --- | ------------------- | -------- |
| 2⁴       | 16            | 18        | Sim        | 1   | `18 - 16`           | 2        |
| 2³       | 8             | 2         | Não        | 0   | `2` (mantém)        | 2        |
| 2²       | 4             | 2         | Não        | 0   | `2` (mantém)        | 2        |
| 2¹       | 2             | 2         | Sim        | 1   | `2 - 2`             | 0        |
| 2⁰       | 1             | 0         | Não        | 0   | `0` (mantém)        | 0        |


---

### ✅ Resultado final:

```python
Bin = [1, 0, 0, 1, 0]
```

## ⚡️ Versão rápida em Python: usando `bin()`

Python já tem uma função nativa que faz a conversão de decimal para binário:

```python
N = 15
binario = bin(N)
print(binario)
```

### 🧾 Saída:

```
0b1111
```

---

### ✅ O que significa esse `0b`?

* A função `bin` retorna uma **string** com o número binário, **mas com o prefixo `0b`**, indicando que aquilo é uma representação binária.
* O que nos interessa é a parte depois disso: `"1111"`.

---

## 🧼 Removendo o `0b` (forma limpa)

Se você quiser **somente os dígitos binários**, pode fazer assim:

```python
N = 15
binario = bin(N)[2:]
print(binario)
```

### 🧾 Saída:

```
1111
```

---


## ✂️ Como funciona `[2:]`?

`[2:]` é um exemplo de **fatiamento de string** no Python.
Em inglês, isso é chamado de **slicing**.

### ✅ Regra:

```python
texto[init:end]
```

* `init`: a posição onde queremos começar.
* `end`: a posição onde queremos parar (não incluída).
* Se omitido o `end`, ele vai até o final da string.

---

### Então, o que faz `bin(N)[2:]`?

ex:

```python
bin(15) → '0b1111'
         012345
```

As posições (índices) dos caracteres são:

| Índice | Caractere |
| ------ | --------- |
| 0      | '0'       |
| 1      | 'b'       |
| 2      | '1'       |
| 3      | '1'       |
| 4      | '1'       |
| 5      | '1'       |

→ `bin(15)[2:]` **começa do índice 2**, ou seja, pega apenas `"1111"`.

---

### 🎯 Resultado final:

```python
bin(15)[2:] → '1111'
bin(18)[2:] → '10010'
```

## 📦 Função rápida personalizada

Você pode criar uma função simples que faz isso para qualquer número:

```python
def decimal_para_binario(N):
    return bin(N)[2:]

# Exemplo:
print(decimal_para_binario(18))  # Saída: 10010
print(decimal_para_binario(31))  # Saída: 11111
```
