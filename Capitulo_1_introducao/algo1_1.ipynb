Transforma√ß√£o de um n√∫mero inteiro N em base decimal para a base bin√°ria




````python
import math

def decToBin(N): # Define uma fun√ß√£o para converter decimal para bin√°rio.

  # Identifica√ß√£o da maior pot√™ncia de 2 contida em N
  # Determinar N tal que 2^n <= N < 2^(n+1)
  n=0
  pot=1 #2^0
  while pot <= N:
    n+=1
    pot*=2

  pot=pot/2
  n-=1

  # Calculo dos coeficientes ai em base bin√°ria

  Bin = [] #an
  M = N
  for i in range(n+1):
    if M <pot:
      Bin.append(0)
    else:
      Bin.append(1)
      M = M - pot
    pot=pot/2
  return Bin

N=15
Bin = decToBin(N)
print('A representa√ß√£o bin√°ria de '+ str(N) + ' √©:'+ str(Bin))
````

# üßÆ Convertendo um n√∫mero decimal para bin√°rio com Python

Vamos aprender passo a passo como funciona esse c√≥digo Python que converte um n√∫mero decimal (ex: 15) para bin√°rio (como 1111).

---

## üß† Objetivo

Transformar um n√∫mero que usamos no dia a dia (decimal) para a forma como os computadores entendem (bin√°rio).

---

## üí° Conceito b√°sico de bin√°rio

No sistema bin√°rio, usamos **apenas 0 e 1**.
Cada posi√ß√£o representa uma **pot√™ncia de 2**.
Por exemplo, o n√∫mero decimal 15 em bin√°rio √©:

```
2¬≥  2¬≤  2¬π  2‚Å∞
 8 + 4 + 2 + 1 = 15
 ‚Üí   1   1   1   1
```

Portanto, 15 em bin√°rio √© `1111`.

---

## üßæ O c√≥digo Python


```python
import math

def decToBin(N):
    # Descobrir a maior pot√™ncia de 2 menor ou igual a N
    n = 0
    pot = 1  # Come√ßa com 2^0

    while pot <= N:
        n += 1
        pot *= 2  # Multiplica por 2 em cada passo

    pot = pot / 2
    n -= 1

    # Construir a representa√ß√£o bin√°ria
    Bin = []  # Lista que vai guardar os 0s e 1s
    M = N     # N√∫mero original, que ser√° diminu√≠do

    for i in range(n + 1):
        if M < pot:
            Bin.append(0)
        else:
            Bin.append(1)
            M = M - pot
        pot = pot / 2  # Vai para a pr√≥xima pot√™ncia menor

    return Bin

# Testando a fun√ß√£o
N = 15
Bin = decToBin(N)
print('A representa√ß√£o bin√°ria de', N, '√©:', Bin)
```

## üß© Explica√ß√£o do c√≥digo passo a passo

### 1. `import math`

Importa a biblioteca `math`, mas que **n√£o √© usada nesse c√≥digo**.

---

### 2. `def decToBin(N):`

Define uma **fun√ß√£o** chamada `decToBin` que recebe um n√∫mero `N`.

* `def` ‚Üí palavra-chave para criar fun√ß√µes.
* `N` ‚Üí o n√∫mero que queremos converter.

---

### 3. Encontrando a maior pot√™ncia de 2

```python
n = 0
pot = 1

while pot <= N:
    n += 1
    pot *= 2
```

* Esse loop `while` calcula a **maior pot√™ncia de 2 que cabe no n√∫mero N**.
* Exemplo com `N = 15`:

| Repeti√ß√£o | `pot` (2^n)    |
| --------- | -------------- |
| 1         | 1              |
| 2         | 2              |
| 3         | 4              |
| 4         | 8              |
| 5         | 16 (para aqui) |

‚Üí A maior pot√™ncia de 2 que **n√£o passa** de 15 √© `8`, que √© `2^3`.

Depois, fazemos:

```python
pot = pot / 2  # volta para 8
n -= 1         # n vira 3
```

---

### 4. Construindo o n√∫mero bin√°rio

```python
Bin = []
M = N
```

* `Bin` √© a lista onde vamos guardar os d√≠gitos (0 ou 1).
* `M` √© uma c√≥pia de `N`, que vamos diminuir √† medida que usamos pot√™ncias de 2.

---

### 5. Loop para montar os bits

```python
for i in range(n + 1):
    if M < pot:
        Bin.append(0)
    else:
        Bin.append(1)
        M = M - pot
    pot = pot / 2
```

Esse comando significa:

> ‚ÄúRepita os passos abaixo **(n + 1)** vezes‚Äù.

* `n + 1` √© usado porque estamos contando **inclusive a pot√™ncia 2‚Å∞**, que tamb√©m faz parte do n√∫mero bin√°rio.
* Exemplo: Se `n = 3`, o loop ser√° executado com `i = 0, 1, 2, 3` (4 vezes no total).

üí° **`range(n + 1)`** √© como criar uma **sequ√™ncia de n√∫meros de 0 at√© n**.

---

### üîé Vari√°veis usadas

Antes do `for`, j√° temos algumas vari√°veis importantes:

* `M`: √© o n√∫mero que vai ser representado (vai diminuindo conforme usamos pot√™ncias de 2).
* `pot`: √© o valor da pot√™ncia de 2 que estamos testando (come√ßa com a maior e vai diminuindo).
* `Bin`: √© a lista onde vamos guardar os **0s e 1s** da resposta final.

---

## üëá Passos dentro do loop

Agora vamos ver o que acontece **em cada repeti√ß√£o** do `for`.

### üîπ `if M < pot:`

Essa linha faz uma **pergunta**:

> ‚ÄúA pot√™ncia atual de 2 cabe dentro de `M`?‚Äù

Se **n√£o cabe** (ou seja, `M` √© menor que `pot`), colocamos `0`:

```python

```

---

### üîπ `else:` ‚Üí quando a pot√™ncia **cabe**

Se a pot√™ncia cabe no valor que ainda temos (`M`), fazemos 2 coisas:

```python

M = M - pot      # Subtrai essa pot√™ncia do valor restante
```

Ou seja: usamos essa ‚Äúpot√™ncia de 2‚Äù e atualizamos o quanto ainda falta.

---

### üîª `pot = pot / 2`

No final de cada repeti√ß√£o, dividimos `pot` por 2 para testar a **pr√≥xima menor pot√™ncia de 2**.

Exemplo: se `pot = 8`, depois ser√° `4`, depois `2`, depois `1`, etc.

---

## üîö Resumo final da l√≥gica

1. Come√ßamos com o valor original (ex: 15).
2. Verificamos se a pot√™ncia de 2 atual **cabe** dentro desse valor.
3. Se sim, colocamos `1` e subtra√≠mos.
4. Se n√£o, colocamos `0`.
5. Repetimos para a pr√≥xima pot√™ncia menor (dividindo por 2).
6. No final, temos uma lista com 0s e 1s representando o n√∫mero bin√°rio.

## ‚ö°Ô∏è Vers√£o r√°pida em Python: usando `bin()`

Python j√° tem uma fun√ß√£o nativa que faz a convers√£o de decimal para bin√°rio:

```python

print(binario)
```

### üßæ Sa√≠da:

```

```

---

### ‚úÖ O que significa esse `0b`?

* A fun√ß√£o `bin` retorna uma **string** com o n√∫mero bin√°rio, **mas com o prefixo `0b`**, indicando que aquilo √© uma representa√ß√£o bin√°ria.
* O que nos interessa √© a parte depois disso: `"1111"`.

---

## üßº Removendo o `0b` (forma limpa)

Se voc√™ quiser **somente os d√≠gitos bin√°rios**, pode fazer assim:

```python

```

### üßæ Sa√≠da:

```

```

---


## ‚úÇÔ∏è Como funciona `[2:]`?

`[2:]` √© um exemplo de **fatiamento de string** no Python.
Em ingl√™s, isso √© chamado de **slicing**.

### ‚úÖ Regra:

```python

```

* `init`: a posi√ß√£o onde queremos come√ßar.
* `end`: a posi√ß√£o onde queremos parar (n√£o inclu√≠da).
* Se omitido o `end`, ele vai at√© o final da string.

---

### Ent√£o, o que faz `bin(N)[2:]`?

ex:

```python

         012345
```

As posi√ß√µes (√≠ndices) dos caracteres s√£o:

| √çndice | Caractere |
| ------ | --------- |
| 0      | '0'       |
| 1      | 'b'       |
| 2      | '1'       |
| 3      | '1'       |
| 4      | '1'       |
| 5      | '1'       |

‚Üí `bin(15)[2:]` **come√ßa do √≠ndice 2**, ou seja, pega apenas `"1111"`.

---

### üéØ Resultado final:

```python
bin(15)[2:] ‚Üí '1111'
bin(18)[2:] ‚Üí '10010'
```
