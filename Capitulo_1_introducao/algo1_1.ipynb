Transformação de um número inteiro N em base decimal usando int(x)


```python
import math

def decToBin(N):

  # Identificação da maior potência de 2 contida em N
  # Determinar N tal que 2^n <= N < 2^(n+1)
  n=0
  pot=1 #2^0
  while pot <= N:
    n+=1
    pot*=2

  pot=pot/2
  n-=1

  # Calculo dos coeficientes ai em base binária

  Bin = [] #an
  M = N
  for i in range(n+1):
    if M <pot:
      Bin.append(0)
    else:
      Bin.append(1)
      M = M - pot
    pot=pot/2
  return Bin

N=15
Bin = decToBin(N)
print('A representação binária de '+ str(N) + ' é:'+ str(Bin))

```
