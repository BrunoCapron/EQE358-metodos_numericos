Representação de um número fracionário alpha entre 0 e 1


```python
def fracToBin(alpha,Ndig):

  #Determinação dos coeficientes bi em base binária
  Bin = []
  M = 2*alpha
  for i in range(1,Ndig+1):
    P= int(M)
    M = 2*(M-P)
    Bin.append(P)

  return Bin

Ndig= 5
alpha = 0.7
Bin = fracToBin(alpha,Ndig)
print('A representação binárioa de '+ str(alpha) + ' é: 0,'+ str(Bin))


```
