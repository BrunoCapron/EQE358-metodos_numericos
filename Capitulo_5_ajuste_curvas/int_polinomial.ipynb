Interpolação polinomial

```python
import numpy as np
import math

#função para determinar os coeficientes do polinomio de Newton
def interpolacao_polinomial(x, y):
  n=len(x) #número de pontos

  #construção da tabela para o calculo dos coeficientes
  A=np.zeros((n,n))
  for i in range(n):
    A[i,0] = y[i]
  for i in range(1,n):
    for j in range(1,i+1):
      A[i,j] = (A[i,j-1]-A[i-1,j-1])/(x[i]-x[i-j])
  ai = np.diag(A) #os coeficientes do polinomio são os elementos na diagonal de A
  return ai

def apro_inter_polinomial(x, ai, x_new):
  n=len(ai)
  apro =ai[0]
  for i in range(n):
    prod = 1
    for j in range(i):
      prod = prod*(x_new-x[j])
    apro = apro + prod*ai[i]
  return apro

x = np.array([1,4,5,6])
y = np.array([math.log(1),math.log(4),math.log(5),math.log(6)])

ai = interpolacao_polinomial(x, y)
apro = apro_inter_polinomial(x, ai, 2)
print(apro)
```
