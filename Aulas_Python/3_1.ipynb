resolução de equação algébrica não-linear por método direto


```python
import random
import math

# definição de f(x)
def myfunc(x):

  return 1 - math.exp(x)

# resolução de f(x) = 0 por método direto
def met_direto(myfunc, a, b, kmax, epsilon, delta, metodo):

  k = 0
  delta = b-a
  fa= myfunc(a)
  fb= myfunc(b)
  y= fa

  while k < kmax and (delta > epsilon or abs(y) > delta):
    if metodo == 'bisseccao':
      lambd = 1/2
    else:
      lambd = random.random()

    x = a + lambd*(b-a)
    y = myfunc(x)
    if y*fa > 0:
      a = x
      fa = y
    else:
      b = x
      fb = y
    delta = b-a
    k = k+1
  return x, k, delta

a = -2
b = 3
kmax= 1e6
epsilon = 1e-12
delta = 1e-12
x, k, delta = met_direto(myfunc, a, b, kmax, epsilon, delta, 'bisseccao')


print('x= ' + str(x) + ' k= ' + str(k) + ' delta= ' + str(delta))






```
