Implementação do algoritmo numérico de expansão em série de Taylor de exp(x) com a especificação de delta

```python
import math

def expansao_taylor(x,delta):

  y = x
  m=0
  n=0
  T=1
  S=1

  #definição de m
  while abs(y) >= 1:
    m+=1
    y = y/2

  #cálculo de n
  while abs(T/S) > delta:
    n += 1
    T = y/n*T
    S = S + T

  for j in range(1,m+1):
    S = S**2

  return m, n, S

x = -2.5
delta = 1e-4

m, n, S = expansao_taylor(x,delta)
valor_real = math.exp(x)

erro = S - valor_real

print('m = ' + str(m) + ', n = ' + str(n) + ', S = ' + str(S))
print('Valor real = ' + str(valor_real))
print('Erro = ' + str(erro))

```
