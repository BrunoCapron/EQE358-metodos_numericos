Transformação de um número inteiro N em base decimal usando int(x)


```python
import math

def decToBin(N):

  # Identificação da maior potência de 2 contida em N
  # Determinar N tal que 2^n <= N < 2^(n+1)
  n=int(math.log(N,2))

  #Determinação dos coeficientes ai em base binária

  P=N
  Bin=[]
  for i in range(n+1):
    ai = P % 2 # P mod 2 - resto da divisão de P por 2
    Bin.append(ai)
    P = int(P/2)

  return Bin

N=25
Bin = decToBin(N)
Bin.reverse()
print('A representação binárioa de '+ str(N) + ' é:'+ str(Bin))

```
