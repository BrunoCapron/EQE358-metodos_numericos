## Multiplicação de Matrizes

```python
import numpy as np

A = np.array([[1, 0], [0, 1]])
B = np.array([[4, 1], [2, 2]])

M = A*B     #Produto elemento a elemento! Cuidado!!
print(M)
```

```python
M = np.dot(A,B)
print(M)
```

```python
M = np.matmul(A, B)
print(M)
```

```python
M = A @ B
print(M)
```

## Determinantes

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])

D = np.linalg.det(A)
print(D)
```

```python
from numpy.linalg import det

D = det(A)
print(D)
```

```python
import numpy as np

A = np.array([[6,1,-2,3], [0,-2,4,-1],[1,-1,1,-1],[0,-2,2,1]])

D = np.linalg.det(A)
print(D)
```

## Matriz Inversa

```python
import numpy as np
from numpy.linalg import inv

#A = np.array([[1., 2.], [3., 4.]])
A = np.array([[1., -2.,0], [0., 1.,2],[2., 2.,0]])
Ainv = inv(A)

print(Ainv)
```

```python
AandB = np.array([[[1., 2.], [3., 4.]], [[1, 3], [3, 5]]])
inv(AandB)
```

## Autovalores e Autovetores

```python
import numpy as np
from numpy.linalg import eig
```

```python
A = np.array([[0, 2],
              [2, 3]])
w,v=eig(A)

print('Autovalores:')
print(w)
print('Autovetores:')
print(v)
```

```python
a = np.array([[2, 2, 4],
              [1, 3, 5],
              [2, 3, 4]])
w,v=eig(a)

print('Autovalores:')
print(w)
print('Autovetores:')
print(v)
```

```python
a = np.array([[1, 1, 0],
              [0, -2, 1],
              [0, 0, 3]])
w,v=eig(a)
print('Autovalores:')
print(w)
print('Autovetores:')
print(v)
```
